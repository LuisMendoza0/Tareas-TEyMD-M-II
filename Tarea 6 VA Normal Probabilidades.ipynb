# **Tarea Cálculo de probabilidades**
1 - $\mathbb{P}(0<Z<b)$
  - Queremos calcular $\mathbb{P}(0<Z<0.43)$, lo cuál puede realizarse de la siguiente manera:
    * Tablas de área a la derecha: $\mathbb{P}(0<Z<0.43) = 0.1664$

2 - $\mathbb{P}(-b<Z<b)$
  - Queremos calcular $\mathbb{P}(-0.16<Z<0.16)$, lo cuál puede realizarse de la siguiente manera:
    * Tablas de área a la derecha:  $\mathbb{P}(-0.16<Z<0.16) = \mathbb{P}(-0.16<Z<0)+\mathbb{P} (0<Z<0.16) = \mathbb{P}(0<Z<0.16)+\mathbb{P}(0<Z<0.16) = 2\mathbb{P}(0<Z<0.16) = 2(0.0636) = 0.1272$

3 - $\mathbb{P}(Z<-b)$
  - Queremos calcular $\mathbb{P}(Z<-1.94)$, lo cuál puede realizarse de la siguiente manera:
    * Tablas de área a la derecha: $\mathbb{P}(Z<-1.94) = \mathbb{P}(Z<0)+\mathbb{P}(-1.94<Z<0) = \mathbb{P}(Z<0) + \mathbb{P}(0<Z<1.94) = 0.5 - 0.4738 = 0.0262$

4 - $\mathbb{P}(Z>-b)$
  - Queremos calcular $(Z>-0.07)$, lo cuál puede realizarse de la siguiente manera:
    * Tablas de área a la derecha: $\mathbb{P}(Z>-0.07) = \mathbb{P}(-0.07<Z<0) + \mathbb{P}(Z>0) = \mathbb{P}(0<Z<0.07) + \mathbb{P}(Z>0) = 0.0279 + 0.5 = 0.5279$

## Liberías y Paquetes requeridos

```python
import numpy as np
import numpy.random as npr
import scipy.stats as sps
import matplotlib.pyplot as plt
import random

from scipy.stats import norm

import warnings
warnings.filterwarnings("ignore")
```

## Calculando las probabilidades

```python
# P(0<Z<0.43)
Ej1= norm(0,1).cdf(0.43)-norm(0,1).cdf(0)
# P(-0.16<Z<0.16)
Ej2=norm(0,1).cdf(0.16)-norm(0,1).cdf(-0.16)
# P(Z<-1.94)
Ej3=norm(0,1).cdf(-1.94)
# P(Z>-0.07)
Ej4=1-norm(0,1).cdf(-0.07)
print("Resultados Previos Obtenidos")
print("P(0<Z<0.43)=",round(Ej1,4))
print("P(-0.16<Z<0.16)=",round(Ej2,4))
print("P(Z<-1.94)=",round(Ej3,4))
print("P(Z>-0.07)=",round(Ej4,4))

```

## Gráficos respectivos

A continuación presento nuestro primero gráfico para $\mathbb{P}(0<Z<0.43)$

```python
# Creamos una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Generamos un rango de valores x desde -3 hasta 3 con un paso de 0.001 para la curva de distribución
x = np.arange(-2, 5, 0.001)

# Graficamos la función de densidad de probabilidad de la distribución normal con media 0 y desviación estándar 1
ax.plot(x, norm.pdf(x, loc=0, scale=1))

# Establecemos el título del gráfico
ax.set_title("N(0,$1$)")

# Etiquetamos el eje x
ax.set_xlabel('x')

# Etiquetamos el eje y
ax.set_ylabel('pdf(x)')

# Añadimos una cuadrícula al gráfico
ax.grid(True)

# Generamos un rango de valores px desde 0 hasta 0.43 con un paso de 0.001 para el área bajo la curva
px = np.arange(0, 0.43, 0.001)

# Establecemos el límite del eje y para que vaya de 0 a 0.45
ax.set_ylim(0, 0.5)

# Rellenamos el área bajo la curva de la función de densidad de probabilidad para px, con transparencia del 50% y color verde
ax.fill_between(px, norm.pdf(px, loc=0, scale=1), alpha=0.5, color='g')


# Añadimos un texto al gráfico en la posición (0, 0.41) con el valor de la probabilidad redondeado a 4 decimales y tamaño de fuente 20
ax.text(0, 0.41, round(Ej1, 4), fontsize=20)

# Mostramos el gráfico
plt.show()
```

Presento los cuatro gráficos de los ejercicios realizados, considerando la estructura previa en un arreglo para presentar los 4 de manera simultánea.

```python
# Creamos una figura y un conjunto de ejes con 2 filas y 2 columnas
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
  # La figura es un arreglo (digámoslo así, "matricial"),  el parámetro figsize=(10, 10) establece el tamaño de la figura en pulgadas (ancho, alto).

# Graficamos en cada subfigura la función de densidad de probabilidad de la distribución normal con media 0 y desviación estándar 1
  #axs[i,j] hace referencia a la posición de nuestra figura creada

#EJ1
x1=np.arange(-2, 5, 0.001)
axs[0, 0].plot(x1, norm.pdf(x1, loc=0, scale=1))
axs[0, 0].set_title("P(0<Z<0.43)")
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('pdf(x)')
axs[0, 0].grid(True)
axs[0, 0].set_ylim(0, 0.5)
axs[0, 0].fill_between(np.arange(0, 0.43, 0.001), norm.pdf(np.arange(0, 0.43, 0.001), loc=0, scale=1), alpha=0.5, color='g')
axs[0, 0].text(0, 0.41, round(Ej1, 4), fontsize=12)

#EJ2
x2=np.arange(-1, 5, 0.001)
axs[0, 1].plot(x2, norm.pdf(x2, loc=1, scale=0.5))
axs[0, 1].set_title("P(-0.16<Z<0.16)")
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('pdf(x)')
axs[0, 1].grid(True)
axs[0, 1].set_ylim(0, 1)
axs[0, 1].fill_between(np.arange(-0.16, 0.16, 0.001), norm.pdf(np.arange(-0.16, 0.16, 0.001), loc=1, scale=0.5), alpha=0.5, color='b')
axs[0, 1].text(0.3, 0.16, round(Ej2, 4), fontsize=12)

#EJ3
x3=np.arange(-4, 2, 0.001)
axs[1, 0].plot(x3, norm.pdf(x3, loc=-1, scale=0.8))
axs[1, 0].set_title("P(Z<-1.94)")
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('pdf(x)')
axs[1, 0].grid(True)
axs[1, 0].set_ylim(0, 0.6)
axs[1, 0].fill_between(np.arange(-2, -1.94, 0.001), norm.pdf(np.arange(-2, -1.94, 0.001), loc=-1, scale=0.8), alpha=0.5, color='r')
axs[1, 0].text(-1.8, 0.22, round(Ej3, 4), fontsize=12)

#EJ4
x4=np.arange(-1, 2, 0.001)
axs[1, 1].plot(x4, norm.pdf(x4, loc=2, scale=0.3))
axs[1, 1].set_title("P(Z>-0.07)")
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('pdf(x)')
axs[1, 1].grid(True)
axs[1, 1].set_ylim(0, 1.5)
axs[1, 1].fill_between(np.arange(-0.07, 5, 0.001), norm.pdf(np.arange(-0.07, 5, 0.001), loc=2, scale=0.3), alpha=0.5, color='m')
axs[1, 1].text(2.4, 1, round(Ej4, 4), fontsize=12)

# Agregamos un título general a la figura
fig.suptitle('Distribuciones Normales N(0,$1$)')

# Mostramos el gráfico
plt.tight_layout()
plt.show()
```
