## **1. Dibujar la función de distribución acumulada de una variable aleatoria.**
**$$X\sim \text{Binom}(3,0.3) \hspace{0.2cm} \text{y calcular} \hspace{0.2cm} \mathbb{P}(X\leq 2)$$**



Presento la construcción realizada en excel y después la construcción realizada en equipo
image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parámetros de la distribución binomial
n = 3
p = 0.3

# Valores posibles de la variable aleatoria binomial
x = np.arange(0, n+1)

# Función de distribución acumulada (CDF)
cdf = binom.cdf(x, n, p)

# Gráfico
for i in range(n+1):
    plt.hlines(cdf[i], x[i]-0.5, x[i]+0.5, colors='blue')

plt.title('Función de Distribución Acumulada (CDF) de una Binomial(n=3, p=0.3)')
plt.xlabel('Número de éxitos')
plt.ylabel('Probabilidad acumulada')
plt.xticks(x)
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True)
plt.show()
```

Calculando $\mathbb{P}(X\leq 2)$:

\begin{eqnarray}
\mathbb{P}(X\leq 2) & = & \sum_{k=0}^{2} \binom{3}{k}(0.3)^{k}(1-0.3)^{n-k}\\ & = & \binom{3}{0}(0.3)^{0}(0.7)^{3-0} + \binom{3}{1}(0.7)^{1}(0.7)^{3-1} + \binom{3}{2}(0.3)^{2}(0.7)^{3-2} \\ & = & 0.7^3 + (3)(0.7)(0.7)^{2} + (3)(0.3)^{2}(0.7) \\ & \approx & 0.973
\end{eqnarray}




Otra alternativa para calcular $\mathbb{P}(X\leq 2)$ con Python y la librería `scipy.stats`:

```python
from scipy.stats import binom

# Parámetros de la distribución
n = 3
p = 0.3

# Se calcula P(X <= 2)
prob_x_menor_2 = binom.cdf(2, n, p)

print("\n\n\t\t Por lo tanto; Si X ~ Binom(3,0.3), entonces P(X<=2) =",prob_x_menor_2,"\n\n" )
```

## **2. Determinar si la variable aleatoria dada es una v.a. binomial.**

**1.** *$X$ es el número de canicas negras en una muestra de $5$ canicas extraídas al azar y sin reemplazo de una caja que contiene $25$ canicas blancas y $15$ canicas negras.*

**Análisis y Conclusión**

La descripción de la variable aleatoria mencionada nos brinda la siguiente información
  - Canicas Blancas ⚪= 25
  - Canicas Negras ⚫= 15
  - Total de Canicas en la Caja = 40
  - Tamaño de la Muestra (extracción) =5
  - Restricciones $\rightarrow$ Sin reemplazo

Tenemos ciertos aspectos que permitirían concluir que es una variable aleatoria binomial pues:
  - Tenemos una cantidad de "experimentos" (las extracciones) específica que es de 5.
  - Tenemos dos posibles resultados, éxito (canicas blancas) y fracaso (canicas negras).

Sin embargo, donde rompe la construcción de una variable aleatoria binomial es en los "eventos independientes" ya que nos restringe a no tener reemplazos, lo que implica que no tendremos una probabilidad **$p$** de "éxito" específica ya que con cada extracción de la caja, el número de canicas disminuirá.

Por lo tanto esta variable aleatoria **$X$** no será binomial.

**2.** *$X$ es el número de canicas negras en una muestra de 5  canicas extraídas al azar y con reemplazo de una caja que contiene $25$  canicas blancas y $15$  canicas negras.*

**Análisis y Conclusión**

Dada lo anterior podemos notar lo siguiente:

- Hay un número fijo de ensayos, el cual es $n=5$.
- Cada ensayo tiene dos posibles resultados:
    - Obtener una canica negra ⚫ (éxito)
    - Obtener una canica blanca ⚪ (fracaso)
- La probabilidad de éxito $p$ es constante en cada ensayo: dado que las canicas se extraen con reemplazo, la probabilidad de extraer una canica negra se mantiene constante en cada extracción. La probabilidad de extraer una canica negra es $0 < p=\frac{15}{40} < 1$.
- La extracción con reemplazo garantiza que cada ensayo es independiente del anterior.

De modo que, podemos concluir que $X\sim \text{Binom}\left(5, \frac{15}{40}\right)$.


**3.** *$X$ es el número de votantes a favor de la ley propuesta en una muestra de $1200$ votantes seleccionados aleatoriamente de todo el electorado de un país en el que el $35\%$ de los votantes están a favor de la ley.*

**Análisis y Conclusión**

Tomamos en cuenta las siguientes condiciones,

  - **Número fijo de Votantes $(n)$:** En este caso, el número de ensayos es el número de votantes seleccionados, que es 1200. Esto cumple con la condición de tener un número fijo de ensayos.

  - **Sólo 2 posibles resultados por ensayo:** Cada votante puede estar a favor de la ley (éxito) o no estar a favor de la ley (fracaso). Por lo tanto, tenemos dos posibles resultados para cada ensayo.

  - **Probabilidad constante de éxito $(p)$:** La probabilidad de que un votante esté a favor de la ley es del 35%, o $p=0.35$. A diferencia del ejemplo **1**, esta probabilidad es constante para cada votante seleccionado.

  En virtud de los puntos anteriores, podemos concluir que la variable aleatoria $X$  sigue una distribución binomial. Es decir

  $\begin{equation}
    \hspace{9 cm} X \sim Binom(1200,0.35)
  \end{equation}$






**4.** *$X$ es el número de monedas que coinciden con al menos otra moneda cuando se lanzan cuatro monedas a la vez.*

**Análisis y Conclusión**

Considerando lo anterior, se puede considerar que es una variable aleatoria binomial

  - Ensayos fijos, $n=4$
  - Ensayos independientes

Por el contrario, existen otros aspectos para no considerar esto

  - Dos posibles resultados: No tiene, ya que hay múltiples posibles resultados
  - La probabilidad de éxito es ambigua


  Por lo que se concluye que esta variable aleatoria $X$ no es binomial.

##**3. Investigar 5 aplicaciones del uso de la variable aleatoria Poisson en la vida diaria.**

1. **Los servicios de streaming, plataformas y páginas web**

Las empresas y sitios alojados en línea utilizan la distribución de Poisson para poder predecir la cantidad visitas que reciben cada hora y establecer suficiente ancho de banda para evitar que se colapsen sus plataformas.

2. **Los restaurantes y servicios de comida rápida**

Las cadenas de alimentos y restaurantes emplean la distribución de Poisson para estimar la cantidad de clientes que recibirán cada día para  estructurar la capacidad de lugares y  de personal.

3. **Control de tráfico vehicular en intersecciones**

La distribución de Poisson se utiliza para modelar el número de vehículos que llegan a una intersección en un cierto periodo de tiempo. Esto ayuda en la planificación y gestión del flujo vehicular, el diseño de semáforos y la mejora de la seguridad en las intersecciones.

4. **Centro de Atención al Cliente**


Un centro de llamadas puede usar la distribución de Poisson para estimar el número de llamadas entrantes por hora, que ayudara para la planificación del personal y gestión de recursos.


5. **Eventos Deportivos**


Los organizadores de eventos deportivos emplean la distribución de Poisson para estimar el número de espectadores que llegarán a diferentes entradas del estadio en ciertos intervalos de tiempo. Esto ayuda en la gestión del flujo de personas, la seguridad y la asignación de personal en las entradas.
